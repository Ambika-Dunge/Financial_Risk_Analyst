# Financial Risk Analysis Project

## Objective
Analyze loan data to identify financial risk and default patterns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration for charts
sns.set_style("whitegrid")
%matplotlib inline
```

## 1. Data Loading

```python
df = pd.read_csv('data/loan_data.csv')
print(f"Dataset Shape: {df.shape}")
df.head()
```

## 2. Data Cleaning & Validation

```python
# Check for missing values
print(df.isnull().sum())

# Check data types
print(df.dtypes)

# Basic statistics
df.describe()
```

## 3. Exploratory Data Analysis (EDA)

```python
# Loan Status Counts
print(df['Loan_Status'].value_counts(normalize=True))

plt.figure(figsize=(6,4))
sns.countplot(x='Loan_Status', data=df)
plt.title('Loan Status Distribution')
plt.show()
```

## 4. Financial Risk Analysis

```python
# Income vs Default Risk
plt.figure(figsize=(10,6))
sns.boxplot(x='Loan_Status', y='Income', data=df)
plt.title('Income vs Loan Status')
plt.show()

# Credit Score vs Default Risk
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Credit_Score', hue='Loan_Status', kde=True)
plt.title('Credit Score Distribution by Status')
plt.show()
```

## 5. Insights & Conclusions

- **Credit Score Impact**: Lower credit scores are significantly correlated with defaults.
- **Income Factor**: Lower income brackets show higher default rates.
- **Risk Segment**: Customers with Credit Score < 600 and Income < 30k are high risk.
